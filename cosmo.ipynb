This dataset provides a detailed analysis of cosmetic product sales and customer purchasing behavior within a beauty retail environment. It contains 1000 records and 22 attributes that track sales performance, customer demographics, inventory management, and customer satisfaction.

Feature Description
1. Customer Information

Customer ID
A unique identifier assigned to each customer.

Customer Name
The full name of the customer who purchased the product.

Gender
Represents the gender of the customer.
Categories: Male, Female, Other.

City
The city where the customer made the purchase.

2. Product Information

Product_Code
A unique alphanumeric code assigned to each cosmetic product.

Product Name
The name of the cosmetic or beauty product.

ProductCategory
The category of the product.
Categories: Lipstick, Skincare, Foundation, Eye Makeup, Makeup Accessories.

Brand
The cosmetic brand of the product.
Brands: Lakme, Maybelline, MAC, Sugar Cosmetics, etc.

3.Sales & Revenue Metrics

Price
The original market price of the product.

Discount
The percentage discount applied to the product.

Final Price
The selling price after discount deduction.

Units Sold
The number of units sold during the transaction.

Sales Amount
The total revenue generated from the sale.

Date
The date on which the transaction occurred.

4. Inventory & Operational Details

Stock Available
Indicates the remaining inventory stock available.

Supplier Name
The supplier providing the cosmetic product.

Manufacture Date
The production date of the product.

Expiry Date
The expiration date of the cosmetic product.

Sales_Channel
The platform used for the sale.
Categories: Online Store, Retail Store.

Payment Method
The payment option used by the customer.
Categories: Cash, UPI, Credit Card, Debit Card.

5.  Customer Satisfaction Metrics

Customer Rating
The customer feedback rating based on product quality and satisfaction.
Range: 1 to 5.

LABEL ENCODE

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['Order_ID'] = le.fit_transform(df['Order_ID'])
df['Date'] = le.fit_transform(df['Date'])
df['Customer_ID'] = le.fit_transform(df['Customer_ID'])
df['Customer_Name'] = le.fit_transform(df['Customer_Name'])
df['Gender'] = le.fit_transform(df['Gender'])
df['City'] = le.fit_transform(df['City'])
df['Product_Code'] = le.fit_transform(df['Product_Code'])
df['Product_Name'] = le.fit_transform(df['Product_Name'])
df['Product_Category'] = le.fit_transform(df['Product_Category'])
df['Brand'] = le.fit_transform(df['Brand'])
df['Payment_Method'] = le.fit_transform(df['Payment_Method'])
df['Sales_Channel'] = le.fit_transform(df['Sales_Channel'])
df['Supplier_Name'] = le.fit_transform(df['Supplier_Name'])
df['Manufacture_Date'] = le.fit_transform(df['Manufacture_Date'])
df['Expiry_Date'] = le.fit_transform(df['Expiry_Date'])
print(df)
```

This code uses Label Encoding from scikit-learn to convert categorical text data into numerical values. It transforms columns such as customer names, gender, city, product details, brand, and payment method into encoded numbers so the dataset can be used for machine learning and data analysis. This process is part of Data Preprocessing in Python.


```python
x=df.drop(columns=["Sales_Amount"],axis=1)
print(x)

y=df["Sales_Amount"]
print(y)
```

This code divides the dataset into input and output variables for machine learning. All columns except `Sales_Amount` are stored in `X` as input features, while the `Sales_Amount` column is stored in `Y` as the target variable used for prediction and analysis.


train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=42)
print(f"x_train{x_train}")
print(f"x_test{x_test}")
print(f"y_train{y_train}")
print(f"y_test{y_test}")
```

This code splits the dataset into training data and testing data using `train_test_split()`. Here, 80% of the data is used for training the machine learning model, and 20% is used for testing the model’s performance. The `random_state=42` ensures consistent results each time the code runs.


                                                             MODEL BULIDING

LINEAR REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
predictions=model.predict(x_test)
accuracy=model.score(x_test,y_test)
print(f"predictions{predictions}")
print(f"accuracy {accuracy}")
```

This code trains a Linear Regression model to predict `Sales_Amount`. The model achieved an accuracy score of 90.78%, showing good prediction performance on the dataset.


```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test,predictions)

print(f"MSE: {mse}")
print(f"R2 Score: {r2}")
```

This code evaluates the performance of the Linear Regression model using MSE and R² Score. The model achieved an R² Score of 0.9078, which indicates very good prediction accuracy. The Mean Squared Error (MSE) value of 454041.40 shows the average prediction error between actual and predicted sales values.


DECISION TREE MODEL

```python
from sklearn.tree import DecisionTreeRegressor  

model = DecisionTreeRegressor()                 
model.fit(x_train, y_train)
predictions = model.predict(x_test)
accuracy = model.score(x_test, y_test)

print(f"predictions {predictions}")
print(f"accuracy {accuracy}")
```

This code uses the Decision Tree Regressor algorithm to predict `Sales_Amount`. The model achieved an accuracy score of 97.18%, indicating excellent prediction performance and better accuracy compared to the Linear Regression model.
